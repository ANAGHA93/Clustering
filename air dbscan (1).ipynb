```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the file
df = pd.read_csv("EastWestAirlines.csv")
```

```python
df.head()
```

```python
data=df.drop(columns={"ID#","Award?"})
```

```python
data
```

```python
array=data.values
```

```python
array
```

```python
#Standardizing the data using Standardscaler func
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
#Assigning the eps value and min points and creating the model and forming clusters
dbscan = DBSCAN(eps=5, min_samples=11)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.Series(dbscan.labels_)
```

```python
cl
```

```python
df['cluster']=cl
```

```python
df
```

```python
df.iloc[:,1:].groupby(df.cluster).mean()
```
